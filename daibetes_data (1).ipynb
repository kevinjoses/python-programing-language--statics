```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["Age"],bins=20,kde=True)
plt.title("diabetes classification.csv")
plt.show()
print(df["Age"].describe())
sns.kdeplot(df["Age"],shade=True)
plt.title("diabetes classification.csv")
plt.show()
```

```python
sns.histplot(df["BMI"],bins=20,kde=True)
plt.title("diabetes classification.csv")
plt.show()
print(df["BMI"].describe())
sns.kdeplot(df["BMI"],shade=True)
plt.title("KDE Plot of BMI")
plt.show()
```

```python
sns.histplot(df["FBS"],bins=20,kde=True)
plt.title("diabetes classification.csv")
plt.show()
print(df["FBS"].describe())
sns.kdeplot(df["FBS"],shade=True)
plt.title("KDE plot of FBS")
plt.show()
```

```python
sns.histplot(df["HbA1c"],bins=20,kde=True)
plt.title("diabetes classification.csv")
plt.show()
print(df["HbA1c"].describe())
sns.kdeplot(df["HbA1c"],shade=True)
plt.title("KDE plot of HbA1c")
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x=df["Family History of Diabetes"])
plt.xticks(rotation=40)
plt.title("Family History of Diabetes")
plt.show()
print(df['Family History of Diabetes'].describe())
```

```python
for i in df.columns:
    plt.figure(figsize=(20,5))
    sns.countplot(x=df[i])
    plt.xticks(rotation=40)
    plt.title(i)
    plt.show()
```
