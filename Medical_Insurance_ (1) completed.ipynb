```python
sns.boxplot(df["age"])
plt.title("age")
plt.show()
```

```python
sns.boxplot(df["bmi"])
plt.title("bmi")
plt.show()
```

```python
Q1=df['bmi'].quantile(0.25)
Q3=df['bmi'].quantile(0.75)
IQR=Q3-Q1

Minimum=Q1-1.5*IQR
Maximum=Q3+1.5*IQR

df['bmi']=np.where(df['bmi']>Maximum,Maximum,np.where(df['bmi']<Minimum,Minimum,df['bmi']))
```

```python
sns.boxplot(df["bmi"])
plt.title("bmi")
plt.show()
```

```python
sns.boxplot(df["charges"])
plt.title("charges")
plt.show()
```

```python
Q1=df['charges'].quantile(0.25)
Q3=df['charges'].quantile(0.75)
IQR=Q3-Q1

Minimum=Q1-1.5*IQR
Maximum=Q3+1.5*IQR

df['charges']=np.where(df['charges']>Maximum,Maximum,np.where(df['charges']<Minimum,Minimum,df['charges']))
```

```python
sns.boxplot(df["charges"])
plt.title("charges")
plt.show()
```
